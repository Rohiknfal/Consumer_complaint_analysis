```python


# Function to classify sentiment into a range (1 to 5 for both positive and negative)
def get_sentiment_range(text):
    blob = TextBlob(text)
    polarity = blob.sentiment.polarity
    
    if polarity > 0:
       
        if polarity >= 0.6:
            return -1  
        elif polarity >= 0.4:
            return -2
        elif polarity >= 0.2:
            return-3
        elif polarity > 0:
            return -3 
    elif polarity == 0:
        return -4  
    else:
        
        if polarity <= -0.8:
            return -5 
        elif polarity <= -0.6:
            return -5
        elif polarity <= -0.4:
            return -5
        elif polarity <=0.2:
            return -5  

# Apply the function to classify sentiment into a range
df['sentiment_range'] =x.apply(get_sentiment_range)

df['sentiment_range'] 
```

```python
from sklearn.cluster import KMeans
import seaborn as sns
import matplotlib.pyplot as plt

# Features for Clustering
features = ['Complaint_Type',
'Complaint_Status',
'Resolution_Time_Days',
'Mode_of complaint ',
'Platform_purchased']
X = df[features]

wssse_list = []

# Loop through a range of k values (e.g., from 2 to 10)
for k in range(2, 11):  # Trying k = 2 to k = 10 clusters
    kmeans = KMeans(n_clusters=k, random_state=1)
    kmeans.fit(X)  # Fit the model
    wssse = kmeans.inertia_  # Get the WSSSE (inertia_)
    wssse_list.append(wssse)  # Store the WSSSE value
    print(f"Clusters: {k}, WSSSE: {wssse}")

# Plotting the elbow chart
plt.figure(figsize=(8, 6))
plt.plot(range(2, 11), wssse_list, marker='o', linestyle='-', color='b')
plt.title("Elbow Method For Optimal k")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("WSSSE (Inertia)")
plt.xticks(range(2, 11))  # X-axis labels for different k values
plt.grid(True)
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import seaborn as sns

# Features for Clustering
features = ['Satisfaction_Score','Complaint_Type','Customer_Loyalty_Score','Purchase frequency ',
             'Complaint_Status']

X = df[features]

# Standardizing the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Determine the optimal number of clusters using the Elbow Method (already done in your previous code)
# Assuming you found the optimal k after running the elbow method, let's say k = 3 for this example.
optimal_k = 3

# Fit the KMeans model with the optimal k
kmeans = KMeans(n_clusters=optimal_k, random_state=1)
kmeans.fit(X_scaled)

# Getting the cluster labels
df['Cluster'] = kmeans.labels_

# Calculate the silhouette score
sil_score = silhouette_score(X_scaled, kmeans.labels_)
print(f"Silhouette Score for k={optimal_k}: {sil_score}")

# Optionally, you can visualize the clusters in a 2D scatter plot if you want to inspect the clusters
# For this, we'll reduce dimensions using PCA (Principal Component Analysis)

from sklearn.decomposition import PCA

# Reduce data to 2 dimensions using PCA for visualization
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Plotting the clusters
plt.figure(figsize=(8, 6))
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=df['Cluster'], palette='viridis', s=100, alpha=0.7)
plt.title(f"KMeans Clusters with k={optimal_k}")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.legend(title='Cluster')
plt.show()
```
